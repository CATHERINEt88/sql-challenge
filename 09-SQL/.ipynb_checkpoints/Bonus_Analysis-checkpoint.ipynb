```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#connect to sqlserver/database
engine = create_engine('postgresql://postgres:mypassword@localhost:5432/sql-challenge')
connection = engine.connect()
```

```python
salaries = pd.read_sql("select * from salaries", connection)
salaries.head()
```

```python
employees_dat = pd.read_sql("select * from employees", connection)
employees_dat.head()
```

```python
titles = pd.read_sql("Select * from titles", connection)
titles.head()
```

```python
emp_sal = pd.merge(employees_dat, salaries, on = "emp_no")
emp_sal
```

```python
merge_df = pd.merge(emp_sal, titles, left_on="emp_title_id", right_on = "title_id")
merge_df
```

```python
new_table = merge_df[["title","salary"]]
new_table.head()
```

```python
# group by the title and the average salary
average_salary = new_table.groupby("title")["salary"]
average_salary = round(average_salary.mean(),0)
average_salary = average_salary.reset_index()
average_salary
```

```python
##create histogram to visualise the most common salary ranges for employees
new_table.hist(column = "salary", color = "b")
plt.title("Average Salary Distribution")
plt.xlabel("Salary Range $")
plt.ylabel("Salary Distribution")
plt.tight_layout()
plt.show()

plt.savefig("histogram.png")
```

```python
#create bar chart of average salary by title
x_axis = average_salary["title"]
y_axis = average_salary["salary"]
plt.bar(x_axis, y_axis, color = 'red')
plt.xticks(x_axis, rotation = 70)
plt.title("Avg Salary for each Title")
plt.xlabel("Position Title")
plt.ylabel("Avg Salary Range $")
plt.tight_layout()
plt.show()

plt.savefig("bar_chart.png")
```
